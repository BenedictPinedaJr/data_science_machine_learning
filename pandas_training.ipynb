## Manipulating Data

```python
new_cars_data["brand"].str.lower()
```

```python
new_cars_data["brand"] = new_cars_data["brand"].str.upper()
new_cars_data
```

```python
missing_cars_data = pd.read_csv("data/usa_cars_data_missing.csv")
missing_cars_data
```

```python
# Fill missing values of specified data frame column
missing_cars_data["brand"].fillna("chevrolet", inplace=True)
```

```python
missing_cars_data["model"].fillna("2015", inplace=True)
```

```python
missing_cars_data["year"].fillna(2018, inplace=True)
```

```python
missing_cars_data
```

```python
# Remove rows with missing values
drop_cars_data = pd.read_csv("data/usa_cars_data_missing.csv")
drop_cars_data.dropna(inplace=True)
drop_cars_data
```

```python
# Column from series
seats_column = pd.Series([5,4,5,5,4])

# New column called seats
new_cars_data["Seats"] = seats_column
new_cars_data
```

```python
new_cars_data["Seats"].fillna(5,inplace=True)
new_cars_data
```

```python
# Column from Python list
import random as rd

# Python list should be the same length with data frame row
fuel_economy = [(rd.random() * 10) for x in range(2499)]
new_cars_data["Fuel per 100KM"] = fuel_economy
new_cars_data
```

```python
new_cars_data["Price x Fuel"] = new_cars_data["price"] * new_cars_data["Fuel per 100KM"]
new_cars_data
```

```python
# Create column from a single value
new_cars_data["No. of Wheels"] = 4
new_cars_data
```

```python
# Create column from a single value
new_cars_data["Passed Road Safety"] = True
new_cars_data
```

```python
# Remove a column
new_cars_data.drop("Price x Fuel", axis="columns", inplace=True)
new_cars_data
```

```python
# Shuffle order of data of dataframe
shuffled_cars_data = new_cars_data.sample(frac=1)
shuffled_cars_data
```

```python
# Shuffle only 20% of dataframe
shuffled_cars_data.sample(frac=0.2)
```

```python
# Reset order of data in dataframe
shuffled_cars_data.reset_index(drop=True,inplace=True)
shuffled_cars_data
```

```python
# Use apply to use a function on dataframe
new_cars_data["price"] = new_cars_data["price"].apply(lambda x : x / 1000)
new_cars_data
```
